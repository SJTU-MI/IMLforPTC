```python
import pandas as pd
import numpy as np
import umap
import umap.plot
import matplotlib.cm as cm
from matplotlib.lines import Line2D
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import matplotlib.patches as patches
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
import pandas as pd
path='D:\\MLforPTC\\02optimized&graph_descriptors\\Similarity_Analysis\\'
file='embed_PI1M.pkl'
dataframe = pd.read_pickle(path+file)
```

```python
dataframe
```

```python
y=dataframe["Class"]
x1=dataframe["x1"]
x2=dataframe["x2"]
```

```python
methods = ['UMAP']
class_names = ["Polyolefins",
"Polystyrenes",
"Polyvinyls",
"Polyacrylics",
"Polyhalo-olefins",
"Polydienes",
"Polyethers",
"Polysulfides",
"Polyesters",
"Polyamides",
"Polyurethanes",
"Polyureas",
"Polyimides",
"Polyanhydrides",
"Polycarbonates",
"Polyimines",
"Polyketones",
"Polysulfones",
"Polyphenylenes"]
```

```python
legend_font = {'family' : 'Arial', 'weight' : 'normal', 'size': 10}

colormap='tab20'
scale=1
figsize=6
show_legend=True
cmap = cm.get_cmap(colormap)
cols = cmap(np.linspace(0., 1., len(class_names)))[:, :-1]
fig, ax = plt.subplots(figsize=(5, 5))
colors =cols[y-1]
plt.scatter(x1, x2, c=colors, s=scale, alpha=1)
ax.set_xticks([])
ax.set_yticks([])
legend_elements = []
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width, box.height])
plt.legend(bbox_to_anchor=(0, 0), loc=0, borderaxespad=0)
#plt.axis('off')
for label, name in enumerate(class_names):
    legend_elements.append(Line2D(
        [0], [0], marker='o', lw=0, markersize=10,
        color=cols[label], label=name, alpha=0.8))
    ax.legend(prop=legend_font,bbox_to_anchor=(1, 1), loc=0, borderaxespad=0.2,handles=legend_elements)
#plt.legend(bbox_to_anchor=(0, 0), loc=0, borderaxespad=0)
```
