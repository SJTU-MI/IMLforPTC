```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
from matplotlib.lines import Line2D
```

```python
df_desc = pd.read_pickle("D:\\MLforPTC\\04model_accuracy\\diff_descriptor.pkl")
df_desc
```

```python
colors_list = ['#E66F51','#2A9D8E','#F3A261']
colors_list1 = ['#000000','#000000','#000000']
colors_list2 = ['#BFBFBF']
sns.set(style="ticks",font="Arial",palette=colors_list,color_codes=True)
plt.figure(figsize=(12,5)) 
ax=sns.violinplot(x="descriptor", y="r2",hue="model",data=df_desc,scale="count",scale_hue="True",inner=None,linewidth=1,dodge=True,zorder=0 )
ax=sns.stripplot(x="descriptor", y="r2",hue="model",data=df_desc,palette=colors_list2,size=7,dodge=True,zorder=1 )
ax = sns.pointplot(x="descriptor", y="r2",hue="model",data=df_desc, ci='sd',palette=colors_list1,join=False,scale=1.5,capsize=.15,dodge=0.52,zorder=2)
ax.set_xlabel('Fingerprint', fontsize=16) 
ax.set_ylabel('Accuracy R\u00b2', fontsize=16) 
#ax.set_title('picture', fontsize=14)
ax.set_ylim(ymin =0.50, ymax =0.90)
y = MultipleLocator(0.10)  
ax.yaxis.set_major_locator(y)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
#plt.legend(labels=["RF","Xgboost","MLP"],fontsize=12)
sns.despine(top=False, right=False, left=False, bottom=False)
ax.spines['bottom'].set_linewidth('1.5')
ax.spines['top'].set_linewidth('1.5')
ax.spines['right'].set_linewidth('1.5')
ax.spines['left'].set_linewidth('1.5')
model_name=["RF","Xgboost","MLP"]
label=[1,2,3]
legend_elements = []
legend_font = {'family' : 'Arial', 'weight' : 'normal', 'size': 14}
for label, name in enumerate(model_name):
    legend_elements.append(Line2D(
        [0], [0], marker='s', lw=0, markersize=10,
        color=colors_list[label], label=name, alpha=0.8))
    ax.legend(prop=legend_font,loc="best",borderaxespad=0.2,handles=legend_elements)
plt.tight_layout()
# show
plt.show()
```
